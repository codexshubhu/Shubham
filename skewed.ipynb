```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import boxcox
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_excel('Iris.xlsx')


print("Dataset Preview:")
print(df.head())


numeric_cols = [
    'SepalLengthCm',
    'SepalWidthCm',
    'PetalLengthCm',
    'PetalWidthCm'
]


print("\nSkewness of Features:")
skewness = df[numeric_cols].skew()
print(skewness)


feature = 'PetalLengthCm'


plt.figure(figsize=(6,4))
plt.hist(df[feature], bins=10, edgecolor='black')
plt.title('Before Transformation')
plt.xlabel(feature)
plt.ylabel('Frequency')
plt.show()


df['Log_Transform'] = np.log1p(df[feature])


df['Sqrt_Transform'] = np.sqrt(df[feature])


df['BoxCox_Transform'], lam = boxcox(df[feature])

print("\nBox-Cox Lambda Value:", lam)


plt.figure(figsize=(15,4))


plt.subplot(1,3,1)
plt.hist(df['Log_Transform'], bins=10, edgecolor='black')
plt.title('Log Transformation')
plt.xlabel('Log Transformed')
plt.ylabel('Frequency')


plt.subplot(1,3,2)
plt.hist(df['Sqrt_Transform'], bins=10, edgecolor='black')
plt.title('Square Root Transformation')
plt.xlabel('Sqrt Transformed')


plt.subplot(1,3,3)
plt.hist(df['BoxCox_Transform'], bins=10, edgecolor='black')
plt.title('Box-Cox Transformation')
plt.xlabel('Box-Cox Transformed')

plt.tight_layout()
plt.show()
```
